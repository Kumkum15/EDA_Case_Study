8. Feature Engineering



Create new columns

Original Price: The price before any discount is applied

Total Sales: The total revenue is generated by multiplying the Sales Price by Quantity

Total Profit: The total profit is earned by multiplying the Profit by Quantity

Discount Price: The amount of discount applied, calculated based on the Original Price and Discount

Total Discount: The total discount value for the quantity sold




```python
data['Original Price'] = data['Sales Price'] / (1 - data['Discount'])
data['Total Sales'] = data['Sales Price'] * data['Quantity']
data['Total Profit'] = data['Profit'] * data['Quantity']
data['Discount Price'] = data['Original Price'] * data['Discount']
data['Total Discount'] = data['Discount Price'] * data['Quantity']
```

Create a new column, Shipping Urgency, based on Days to Ship





If Days to Ship is 0, set to "Immediate"

If Days to Ship is between 1 and 3, set to "Urgent"

If Days to Ship is more than 3, set to "Standard"

```python
def label_urgency(days):
  if pd.isna(days):
    return 'unknown'
  elif days == 0:
    return 'Immediate'
  elif 1 <= days <= 3:
    return 'Urgent'
  else:
    return 'Standard'

data['Shipping Urgency'] = data['Days to Ship'].apply(label_urgency)
```

9. Outlier Detection and Handling



Identify and handle outliers in the Sales Price column


```python
# Calculate Q1 and Q3
Q1 = data['Sales Price'].quantile(0.25)
Q3 = data['Sales Price'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = data[(data['Sales Price'] < lower_bound) | (data['Sales Price'] > upper_bound)]

print("Number of outliers:", outliers.shape[0])
print(outliers[['Sales Price']])
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=data['Sales Price'])
plt.title("Boxplot of Sales Column")
plt.show()
```

```python
data_no_outliers = data[(data['Sales Price'] >= lower_bound) & (data['Sales Price'] <= upper_bound)]
data_no_outliers
```

Create a function called remove_outliers that takes as an argument the dataframe and the column that needs to be searched for outliers. Using the 3 * IQR rule, the function should detect and remove the outliers to return the cleaned dataframe


```python
def remove_outliers(df, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    # Using 3*IQR for a stricter outlier removal
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    # Filter out outliers
    cleaned_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)].copy()

    return cleaned_data
```

Now use the function to detect outliers and remove outliers from the Sales Price and Profit columns


```python
# we have removed outliers from sales price column above only so we will remove for profit now using the function

profit_outliers_removed = remove_outliers(data, 'Profit')
profit_outliers_removed
```

10. Customer Segmentation and Analysis



What is a Quintile? Quintiles are a statistical way of dividing data into five equal parts, each representing 20% of the data. For example, customers in the top quintile (Q5) represent the top 20% of sales or profit


Calculate Customer Sales Quintile and Customer Profit Quintile based on total sales and total profit per Customer ID


```python


# Step 1: Group by Customer ID to calculate total Sales and Profit
customer_summary = data.groupby('Customer ID')[['Sales Price', 'Profit']].sum().reset_index()
customer_summary.rename(columns={'Sales Price': 'Total Sales', 'Profit': 'Total Profit'}, inplace=True)

# Step 2: Create Sales Quintile
customer_summary['Sales Quintile'] = pd.qcut(customer_summary['Total Sales'], q=5, labels=[1, 2, 3, 4, 5])

# Step 3: Create Profit Quintile
customer_summary['Profit Quintile'] = pd.qcut(customer_summary['Total Profit'], q=5, labels=[1, 2, 3, 4, 5])

# Step 4: View result
print(customer_summary.head())
```

Create a cross-grid (cross-tabulation) based on these two quintiles to analyze the relationship between customer sales and profitability


```python
# Cross-tabulate Sales Quintile vs Profit Quintile
cross_grid = pd.crosstab(
    customer_summary['Sales Quintile'],
    customer_summary['Profit Quintile'],
    rownames=['Sales Quintile'],
    colnames=['Profit Quintile'],
    margins=True  # Adds totals
)

# Display the result
print(cross_grid)
```
